## Exercise 1
**Read Total profit of all months and show it using a line plot**

The line plot graph should look like this.

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
facecream = data['facecream'].tolist()
toothpaste = data['toothpaste'].tolist()
facewash = data["facewash"].tolist()
bathingsoap= data["bathingsoap"].tolist()
shampoo = data["shampoo"].tolist()
moisturizer = data["moisturizer"].tolist()
plt.plot(Month,facecream,"c.-",label="face cream sales data")
plt.plot(Month,bathingsoap,"g.-",label="bathingsoap sales data")
plt.plot(Month,shampoo,"m.-",label="shampoo sales data")
plt.plot(Month,moisturizer,"b.-",label="moisturizer sales data")
plt.plot(Month,facewash,"r.-",label="face wash sales data")
plt.plot(Month,toothpaste,"k.-",label="toothpaste sales data")
plt.title("sale data")
plt.xlabel("Month number")
plt.ylabel("sales units in number")
plt.legend(loc="upper left")
plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
my_labels = 'toothpaste', 'facewash', 'facecream', 'moisturizer', 'shampoo', 'bathingsoap'
t=0
fw=0
fc=0
m=0
s=0
b=0 
#total profit for each product
for i in toothpaste:
    t=t+i
for i in facewash:
    fw=fw+i
for i in facecream:
    fc=fc+i
for i in moisturizer:
    m=m+i
for i in shampoo:
    s=s+i
for i in bathingsoap:
    b=b+i
sizes = [t, fw, fc, m,s,b]
plt.pie(sizes, labels = my_labels, autopct='%1.1f%%')
plt.title("Sales data")
plt.legend(my_labels,loc="lower left")
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
plt.subplot(2,1,1)
plt.plot(Month,bathingsoap,"ro-")
plt.ylabel("Sales units in number")
plt.title("sales data of a Bathingsoap")

plt.subplot(2, 1, 2)
plt.plot(Month,facewash,"bo-")
plt.xlabel("Month number")
plt.title("sales data of a facewash")

plt.subplots_adjust(hspace=0.4)
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
plt.stackplot(Month,facewash,facecream,toothpaste,bathingsoap ,shampoo ,moisturizer, 
              labels=["facewash","facecream","toothpaste","bathingsoap ","shampoo","moisturizer"],
              colors=["r","b","c","k","g","y"])
plt.legend(loc="upper left")
plt.xlabel("Month number")
plt.ylabel("Sales units in Number")
plt.title("all product sales data")
plt.show()
```
